## This notebook reads in the WA dataset and aggregates the data then saves back to csv


```python
import pandas as pd
import numpy as np
import datetime
```

```python
wa_df = pd.read_csv("data/WA_Electric_Vehicle_Title_and_Registration_Activity.csv")
```

```python
wa_df.head()
```

```python
#we only want to count those data records that are "original title".  This way we do not count the same car multiple times when its registrations gets updated.
wa_original_title = wa_df.loc[wa_df["Transaction Type"]== "Original Title"]
```

```python
wa_original_title.head()
```

```python
wa_original_title.dtypes
```

```python
#create a date column
wa_original_title["date"] = pd.to_datetime(wa_original_title['DOL Transaction Date'])

```

```python
#convert the date column to MonthYear format
wa_original_title["MonthYear"] = wa_original_title['date'].dt.to_period('M')
```

```python
wa_original_title.head()
```

```python
#count the number of original titles recorded in each monthyear
wa_count= wa_original_title['MonthYear'].value_counts().rename_axis('MonthYear').to_frame('Number of EV')

wa_count.head()
```

```python
wa_count.to_csv('data/WA_EV_byMonthYear.csv')
```
